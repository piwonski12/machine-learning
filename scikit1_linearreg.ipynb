```python
from sklearn import linear_model
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Salary_dataset.csv')
#dataset link: https://www.kaggle.com/datasets/abhishek14398/salary-dataset-simple-linear-regression
df.head(2)
```

```python
df[df.isna().any(axis=1)]
#check if there are any nans in Dataframe
```

```python
df = df.dropna()
#if yes, get rid of them in any way u want 
```

```python
x = df['YearsExperience'].values
x = x.reshape(-1,1) 
#reshape to get 2d array
```

```python
y = df['Salary'].values
y
```

```python
reg = linear_model.LinearRegression()
reg.fit(x, y)
```

```python
reg.predict([[10]])
#just testing if everything is fine
```

```python
predy = reg.predict(x)
#pred Y to create nice chart
```

```python
plt.figure(figsize=(10,10))
#1000 x 1000 px
plt.plot(x,y, color='blue', linewidth=2, label='Train Data')
plt.plot(x, predy, color = 'green', label='Linear Regression')
plt.xlabel('Years Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
